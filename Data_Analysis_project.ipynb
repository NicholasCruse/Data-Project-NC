Select Options

```python
#a list of columns that are not needed.
cols_to_drop = ['Img', 'Last_Update', 'Developer', 'Genre', 'Na_Sales', 'Jp_Sales', 'Pal_Sales', 'Other_Sales']
#drop the listed columns from the table
Game_Sales = Game_Sales.drop(columns = cols_to_drop)
```

```python
#list the current updated table
list(Game_Sales.columns)
```

```python
#display how many rows and columns that are in the current table
df_shape = Game_Sales.shape
print(f'The DataFrame has {df_shape[0]} rows and {df_shape[1]} columns!')
```

Data Cleaning and Preparation

```python
#drops any row that have missing variable within the filter
Game_Sales_Cleaned = Game_Sales.dropna(subset=['Critic_Score', 'Total_Sales'])
```

```python
#displays how many rows and columns after droping a couple of rows with missing variables
df_shape = Game_Sales_Cleaned.shape
print(f'The DataFrame has {df_shape[0]} rows and {df_shape[1]} columns!')
```

```python
#provides summary of the numeric columns with in the table and rounds it up to two decimanls
Game_Sales_Cleaned.describe().round(2)
```

```python
#give the min, median, max, and mean of the total sales
minT = Game_Sales_Cleaned['Total_Sales'].min()
medianT = Game_Sales_Cleaned['Total_Sales'].median()
maxT = Game_Sales_Cleaned['Total_Sales'].max()
meanT = Game_Sales_Cleaned['Total_Sales'].mean()
print(f"The Total Sales min {minT}, median {medianT}, max {maxT}, and the mean is {meanT}.")
#give the min, median, max, and mean of the Critic Score
minC = Game_Sales_Cleaned['Critic_Score'].min()
medianC = Game_Sales_Cleaned['Critic_Score'].median()
maxC = Game_Sales_Cleaned['Critic_Score'].max()
meanC = Game_Sales_Cleaned['Critic_Score'].mean()
print(f"The Critical score min {minC}, median {medianC}, max {maxC}, and the mean is {meanC}.")
```
